```python

import pandas as pd

class SalesDataAnalyzer:
    def __init__(self):
        self.df = None

    # ================= DATA LOADING =================
    def load_dataset(self):
        try:
            self.df = pd.read_csv("sales_dataset.csv")
            print("Dataset loaded successfully!")
        except Exception as e:
            print("Error loading dataset:", e)
    
    # ================= EXPLORE DATA =================
    def explore_data(self):
        if self.df is None:
            print("Please load the dataset first!")
            return
        
        while True:
            print("\n1. First 5 rows\n2. Last 5 rows\n3. Column names\n4. Data types\n5. Basic info\n6. Exit")
            ch = int(input("Enter your choice: "))
            
            if ch == 1:
                print(self.df.head())
            elif ch == 2:
                print(self.df.tail())
            elif ch == 3:
                print(self.df.columns)
            elif ch == 4:
                print(self.df.dtypes)
            elif ch == 5:
                self.df.info()  
            elif ch == 6:
                print("Exit...")
                break
            else:
                print("Invalid choice!")

analyzer = SalesDataAnalyzer()
analyzer.load_dataset()
analyzer.explore_data()
```

# MISSING DATA

```python
import pandas as pd

class SalesDataAnalyzer:
    def __init__(self):
        self.df = None

    # ================= DATA LOADING =================
    def load_dataset(self):
        try:
            self.df = pd.read_csv("sales_dataset.csv")
            print("Dataset loaded successfully!")
        except Exception as e:
            print("Error loading dataset:", e)

    # ================= MISSING DATA =================        
    def handle_missing_data(self):
        if self.df is None:
            print(" Please load the dataset first!")
            return
        
        while True:
            print("\n1. Display missing rows\n2. Fill with mean\n3. Drop missing rows\n4. Exit")
            ch = int(input("Enter your choice: "))

            if ch == 1:
                print("\nMissing Rows:")
                print(self.df[self.df.isnull().any(axis=1)])

            elif ch == 2:
                self.df.fillna(self.df.mean(numeric_only=True), inplace=True)
                print(" Missing values filled with mean!")

            elif ch == 3:
                self.df.dropna(inplace=True)
                print(" Missing rows dropped!")

            elif ch == 4:
                print("Exit missing data menu...")
                break

            else:
                print(" Invalid choice!")
analyzer = SalesDataAnalyzer()
analyzer.load_dataset()
analyzer.handle_missing_data()
                
```

# VISUALIZATION

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

class SalesDataAnalyzer:
    def __init__(self):
        self.df = None

    # ================= DATA LOADING =================
    def load_dataset(self):
        try:
            self.df = pd.read_csv("sales_dataset.csv")
            print("Dataset loaded successfully!")
        except Exception as e:
            print("Error:", e)
    
    # ================= VISUALIZATION =================
    def visualize(self):
        if self.df is None:
            print(" Please load the dataset first!")
            return
        
        # Convert Date to datetime and extract Year
        if 'Date' in self.df.columns:
            self.df['Date'] = pd.to_datetime(self.df['Date'])
            self.df['Year'] = self.df['Date'].dt.year        
        
        while True:   
            print("""
1. Bar Plot (Sales by Product)
2. Line Plot (Sales by Year)
3. Scatter Plot (Sales vs Year)
4. Pie Chart (Sales by Region)
5. Histogram (Sales Distribution)
6. Stack Plot (Sales by Year)
7. Exit
            """)
            
            ch = int(input("Enter your choice: "))

            if ch == 1:
                self.df.groupby('Product')['Sales'].sum().plot(kind='bar')
                plt.title("Sales by Product")
                plt.show()

            elif ch == 2:
                self.df.groupby('Year')['Sales'].sum().plot(kind='line')
                plt.title("Sales Trend by Year")
                plt.show()

            elif ch == 3:
                plt.scatter(self.df['Year'], self.df['Sales'])
                plt.xlabel("Year")
                plt.ylabel("Sales")
                plt.title("Sales vs Year")
                plt.show()

            elif ch == 4:
                self.df.groupby('Region')['Sales'].sum().plot(kind='pie', autopct='%1.1f%%')
                plt.title("Sales by Region")
                plt.ylabel("")
                plt.show()

            elif ch == 5:
                plt.hist(self.df['Sales'], bins=5)
                plt.title("Sales Distribution")
                plt.show()

            elif ch == 6:
                yearly_sales = self.df.groupby('Year')['Sales'].sum()
                plt.stackplot(yearly_sales.index, yearly_sales.values)
                plt.title("Stack Plot (Sales by Year)")
                plt.show()

            elif ch == 7:
                print("Exit.....")
                break  

            else:
                print(" Invalid choice!")

analyzer = SalesDataAnalyzer()
analyzer.load_dataset()
analyzer.visualize()                
```

# SEABORN

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

class SalesDataAnalyzer:
    def __init__(self):
        self.df = None

    # ================= DATA LOADING =================
    def load_dataset(self):
        try:
            self.df = pd.read_csv("sales_dataset.csv")
            print("Dataset loaded successfully!")
        except Exception as e:
            print("Error:", e)
    
    # ================= SEABORN =================
    def seaborn_visuals(self):
        sns.boxplot(x='Region', y='Sales', data=self.df)
        plt.title("Sales by Region")
        plt.show()

        sns.heatmap(self.df.corr(numeric_only=True), annot=True)
        plt.title("Correlation Heatmap")
        plt.show()
        
analyzer = SalesDataAnalyzer()
analyzer.load_dataset()
analyzer.seaborn_visuals()                
```
